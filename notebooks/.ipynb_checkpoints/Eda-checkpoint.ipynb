# Flight Delay EDA

This notebook explores the 2024 US domestic flight dataset (~7.1M flights, 35 columns) as the foundation for an **arrival-delay prediction system that forecasts before departure** (at scheduling / booking time). Before any modeling we clean and understand the raw data. In this exploratory pass we load the data, quantify missing values, and remove records that cannot carry a delay label, so that later steps work from a clean, well-understood table. The aim is to understand data quality, the distribution of delays, and which fields are usable as predictive features.

## Data cleaning

### Step one: import the data

We load the full 2024 flight file into a pandas DataFrame. `low_memory=False` tells pandas to scan the whole file before inferring column types, which avoids mixed-type warnings on the larger text columns. The resulting table has **7,079,081 rows and 35 columns**, which we confirm with `.shape`.

```python
flights = pd.read_csv("../data/raw//flight_data_2024.csv",low_memory=False)
```

```python
flights.shape
```

```python
flights.dtypes
```

### Step two: null check

We count missing values per column, as both a raw count and a percentage of all rows, then sort by null percentage to see which fields are incomplete. Most columns are fully populated. The missing values are concentrated in flight-timing fields (`dep_time`, `arr_time`, `arr_delay`, `taxi_out`, etc.) and in `cancellation_code`. These nulls are **not random** — they correspond almost entirely to cancelled flights, which never departed or arrived, and to `cancellation_code` being populated only for those cancelled rows.

```python

null_counts = flights.isna().sum()
null_pct = (null_counts / len(flights) * 100).round(2)

null_summary = (
    pd.DataFrame({"null_count": null_counts, "null_pct": null_pct})
    .sort_values("null_pct", ascending=False)
)
null_summary
```

```python
null_summary[null_summary["null_count"] > 0]
```

### Step three: where do the nulls come from?

The null counts sit in clear tiers — ~1.31% for the departure fields, ~1.38% for the arrival fields, ~1.61% for `arr_delay` / `air_time` / `actual_elapsed_time`, and 98.6% for `cancellation_code`. That layered structure is a strong hint that the missing values are **not random**: they line up with **flight status**. A flight can either operate normally, be **cancelled** (never departs or arrives), or be **diverted** (departs but doesn't complete at the scheduled destination). Below we confirm that every null is explained by one of these cases.

```python
# How many flights were cancelled vs diverted?
flights[["cancelled", "diverted"]].sum()
```

```python
# Label each flight by status, then measure the null rate of every column within each group.
# This shows *where* the nulls live: which flight status is responsible for each gap.
status = np.where(flights["cancelled"] == 1, "cancelled",
         np.where(flights["diverted"] == 1, "diverted", "operated"))

null_by_status = flights.isna().groupby(status).mean().round(3).T
null_by_status[null_by_status.sum(axis=1) > 0]   # only columns that actually have nulls
```

**Reading the table.** The three columns are the status groups — **cancelled · diverted · operated** — and each value is the *fraction of flights in that group where the column is null* (0 = never missing, 1 = always missing). Every number has a clean physical meaning:

| column | cancelled | diverted | operated |
|---|---|---|---|
| dep_time | 0.962 | 0.000 | 0.0 |
| dep_delay | 0.965 | 0.000 | 0.0 |
| taxi_out | 0.994 | 0.000 | 0.0 |
| wheels_off | 0.994 | 0.000 | 0.0 |
| wheels_on | 1.000 | 0.088 | 0.0 |
| taxi_in | 1.000 | 0.088 | 0.0 |
| arr_time | 1.000 | 0.088 | 0.0 |
| arr_delay | 1.000 | 1.000 | 0.0 |
| cancellation_code | 0.000 | 1.000 | 1.0 |
| actual_elapsed_time | 1.000 | 1.000 | 0.0 |
| air_time | 1.000 | 1.000 | 0.0 |

- **Operated flights** (right column, all `0.0`) are completely clean — zero missing values in every field. The only null is `cancellation_code` (1.0), which is correct: a code exists only for cancelled flights. These are the usable rows.
- **Cancelled flights** (left column) essentially never happened: departure fields are 96–99% null and every arrival field is 100% null, while `cancellation_code` is always present (`0.000` null). The small gradient — `dep_time` 96.2% null vs `taxi_out`/`wheels_off` 99.4% null — shows a thin slice got a departure time logged but were cancelled before actually taxiing or taking off.
- **Diverted flights** (middle column) depart normally (departure fields `0.000` null) and ~91% land somewhere — the diversion airport — so `wheels_on`/`taxi_in`/`arr_time` are only 8.8% null. But `arr_delay`, `air_time`, and `actual_elapsed_time` are 100% null because they are measured against the *scheduled destination*, which was never reached. `cancellation_code` is null since a diverted flight isn't cancelled.

**Takeaway.** The missingness is entirely structural — there are no stray nulls to impute. The target `arr_delay` is missing for **100% of both cancelled and diverted flights** (~1.6% of all rows), so those rows carry no label and cannot train the model. Dropping them with `flights[flights["arr_delay"].notna()]` catches both cases at once; the remaining operated-only data has essentially no missing values, and `cancelled`, `diverted`, and `cancellation_code` become constant/empty on that subset.

### Step four: keep only flights that actually arrived

Because the nulls are structural — they include the **target itself** (`arr_delay`) — we **drop these rows rather than impute them**. You cannot fabricate a label, and filling features for rows that have no label adds nothing to a supervised model.

Rather than filtering on `cancelled` alone, we filter on the target: keep rows where `arr_delay` is present. Since `arr_delay` is null for 100% of *both* cancelled and diverted flights, this single condition removes both (plus any stray label-less row). For this inflight arrival-delay model, cancelled flights carry no usable information — a flight that never operated has no delay to learn from. (Predicting *cancellations* would be a separate model on a separate dataset, and is out of scope here.)

```python
before = len(flights)
flights = flights[flights["arr_delay"].notna()].copy()

dropped = before - len(flights)
print(f"Dropped {dropped:,} rows ({dropped / before:.2%}) with no arr_delay")
print(f"{len(flights):,} operated flights remain")
flights.shape
```

```python
remaining_nulls = flights.isna().sum()
remaining_nulls[remaining_nulls > 0].sort_values(ascending=False)
```

With cancelled and diverted flights removed, the operated-flight table is essentially null-free. Two expected leftovers:

- **`cancellation_code` is now 100% null** — no cancelled rows remain — and `cancelled` / `diverted` are now constant. All three are uninformative on this subset and belong in the drop-at-modeling list.
- Any tiny residual (e.g. a stray `crs_elapsed_time` or `op_carrier_fl_num`) is a handful of rows and can simply be dropped or inspected individually.

The dataset is now clean and every remaining row has a valid `arr_delay` target, so we can move on to exploring the delay distribution itself.

### Step five: check for negative values

A negative number is only a problem in context:

- **Legitimate:** `dep_delay` and `arr_delay` go negative when a flight departs or arrives *early*. Early flights are completely normal — we must keep these, since clipping them to zero would erase real signal and bias the target.
- **Impossible:** durations and distances (`taxi_out`, `taxi_in`, `air_time`, `actual_elapsed_time`, `crs_elapsed_time`, `distance`) cannot physically be negative. Any negatives there are data errors worth investigating.

We scan **every column in the dataset** in one pass (numeric columns only — text and dates can't be negative), then read off which negatives are legitimate and which, if any, are errors.

```python
# Scan EVERY column in the dataset for negatives in one pass.
# is_numeric_dtype covers float, int, and pandas' nullable Int64; text/date columns -> 0.
neg_counts = flights.apply(
    lambda c: int((c < 0).sum()) if pd.api.types.is_numeric_dtype(c) else 0
)
neg_counts.sort_values(ascending=False)
```

**Result: the data is clean — no impossible negatives.** Every duration, distance, and clock-time column reads `0`. The only negatives are in the two delay columns, and both are entirely legitimate:

- **`arr_delay`: 4,318,559 negatives (~62% of operated flights)** — most flights actually arrive *early*.
- **`dep_delay`: 4,068,725 negatives (~58%)** — most flights depart *early* or on time.

This is expected: airlines pad their schedules, so early arrivals are the norm. We keep these values exactly as they are — a negative delay is real signal, not an error, and clipping it to zero would bias the target.

**Takeaway for modeling:** the target `arr_delay` is centred **below zero** (its median is negative), with a long right tail of genuinely late flights pulling the mean back up toward zero. This skew and heavy right tail are the main things our regression target will have to contend with.

### Step six: convert `fl_date` to datetime

`fl_date` loaded as an object (plain string). Converting it to a true `datetime64` unlocks time-based analysis — pulling out month / day-of-week, resampling, and sorting chronologically. We pass the explicit `%Y-%m-%d` format so pandas parses directly instead of inferring the layout row by row (much faster on ~7M rows).

```python
flights["fl_date"] = pd.to_datetime(flights["fl_date"], format="%Y-%m-%d")
flights["fl_date"].dtype
```

### Step seven: duplicate check

A scheduled flight is uniquely identified by its **date, carrier, flight number, origin, destination, and scheduled departure time**. We check for both fully identical rows and repeated flight keys. If the key count is zero, every row is a distinct flight and our counts and aggregations can be trusted; if not, we'd drop the repeats with `drop_duplicates(subset=flight_key)`.

```python
flight_key = ["fl_date", "op_unique_carrier", "op_carrier_fl_num",
              "origin", "dest", "crs_dep_time"]

dup_rows = flights.duplicated().sum()                    # fully identical rows
dup_keys = flights.duplicated(subset=flight_key).sum()   # same scheduled flight
print(f"Fully duplicated rows: {dup_rows:,}")
print(f"Duplicate flight keys:  {dup_keys:,}")
```

**Result: no duplicates.** Both counts are `0` — no fully identical rows and no repeated flight keys. Every row is a distinct scheduled flight, so counts, rates, and group aggregations can be trusted (no `drop_duplicates` needed).

### Step eight: optimize memory (categoricals)

The text columns repeat the same handful of values millions of times — only a few dozen carriers and a few hundred airports/states across ~7M rows. Casting them to pandas `category` replaces the repeated strings with compact integer codes plus a small lookup table, which cuts memory sharply and makes `groupby` / `value_counts` faster throughout the EDA. We convert every low-cardinality string column and report the reduction.

```python
before_mb = flights.memory_usage(deep=True).sum() / 1024**2

cat_cols = ["op_unique_carrier", "origin", "origin_city_name", "origin_state_nm",
            "dest", "dest_city_name", "dest_state_nm"]
flights[cat_cols] = flights[cat_cols].astype("category")

after_mb = flights.memory_usage(deep=True).sum() / 1024**2
print(f"Memory: {before_mb:,.0f} MB -> {after_mb:,.0f} MB "
      f"({1 - after_mb / before_mb:.0%} smaller)")
flights.dtypes
```

**Result: memory nearly halved** — the frame drops from ~4.3 GB to ~1.8 GB (59% smaller) once the seven text columns become `category`. Combined with `fl_date` now stored as `datetime64`, `groupby` / `value_counts` over carriers, airports, and states will be noticeably faster for the rest of the notebook.

## Exploratory data analysis

The data is now clean — operated flights only (**6,965,267 rows**), correct dtypes, no duplicates. We start with the **target, `arr_delay`**, because its shape decides how we model it. From the negative-value check we already expect it to be centred below zero (most flights arrive early) with a long right tail.

Here we quantify that two ways:
1. **Summary statistics and percentiles** — where the bulk sits and how heavy the tail is.
2. **The distribution** — a histogram, clipped so the long tail doesn't flatten the bulk.

```python

flights["arr_delay"].describe(percentiles=[0.1, 0.5, 0.9, 0.95, 0.99]).round(1)
```

```python
ax = flights["arr_delay"].clip(-60, 180).plot.hist(bins=80, figsize=(9, 4))
ax.axvline(0, color="black", lw=1, ls="--", label="on time")
ax.axvline(15, color="red", lw=1, ls="--", label="delayed (15 min)")
ax.set_xlabel("arrival delay (minutes, clipped to [-60, 180])")
ax.set_ylabel("number of flights")
ax.set_title("Distribution of arrival delay")
ax.legend()
plt.tight_layout()
plt.show()
```

**Findings — the target is heavily right-skewed.**

| stat | value | reading |
|---|---|---|
| count | 6,965,267 | operated flights |
| **mean** | **+7.1 min** | pulled up by the tail |
| **median (50%)** | **−6.0 min** | the *typical* flight arrives 6 min **early** |
| std | 58.0 | wide spread |
| 10% / 90% | −23 / +43 | |
| 95% / 99% | +83 / **+215** | 1-in-100 flights is 3.5 h+ late |
| min / max | −126 / **+3803** | max ≈ 63 h — an extreme tail |

Two things stand out:

1. **Mean (+7.1) ≫ median (−6.0).** The average says "late," but the typical flight is actually *early* — textbook right skew. Most of the mass sits just left of zero, with a long, thin tail of severe delays dragging the mean up. The histogram shows the same: a tall spike around 0 / slightly negative, decaying far to the right.
2. **The tail is extreme.** The 99th percentile is 215 min and the max is 3,803 min (~63 h). These are real (weather / operational meltdowns), but they will dominate a squared-error loss if left unchecked.

**Implications for the regression target:** `arr_delay` is usable directly, but the skew and heavy tail need handling:
- **Cap / winsorize** the target (e.g. at the 99th percentile, or a business cap such as 6 h) so the ~63 h outliers don't dominate the loss.
- Prefer a **robust loss / metric** — Huber or quantile loss over plain MSE, and report **MAE**, which is far less tail-sensitive.
- A plain **log transform won't work directly** — early flights make the target negative; if a transform is needed, use a signed variant like `sign(x)·log1p(|x|)`.

The median-below-zero, long-right-tail shape is the key fact to design the regression around.

### Departure delay: the dominant predictor

`dep_delay` is the single most important *inflight-available* signal: by the time the plane is airborne we already know how late it left, and a late departure mechanically tends to produce a late arrival. We quantify that link three ways — the **correlation**, the **joint density** (`dep_delay` vs `arr_delay`), and the distribution of **`arr_delay − dep_delay`** (how much delay a flight recovers, or adds, in the air). This sets the baseline: how much of the target a single feature already explains.

```python

corr = flights["dep_delay"].corr(flights["arr_delay"])
print(f"Pearson correlation (dep_delay vs arr_delay): {corr:.3f}")
```

```python

fig, ax = plt.subplots(figsize=(7, 6))
hb = ax.hexbin(flights["dep_delay"], flights["arr_delay"],
               gridsize=60, extent=(-30, 180, -60, 200),
               bins="log", cmap="viridis", mincnt=1)
ax.axline((0, 0), slope=1, color="red", lw=1, ls="--", label="arr_delay = dep_delay")
fig.colorbar(hb, ax=ax, label="log10(count)")
ax.set_xlabel("departure delay (min)")
ax.set_ylabel("arrival delay (min)")
ax.set_title("Departure vs arrival delay")
ax.legend()
plt.tight_layout()
plt.show()
```

```python

made_up = flights["arr_delay"] - flights["dep_delay"]
made_up.describe(percentiles=[0.1, 0.5, 0.9]).round(1)
```

```python
ax = made_up.clip(-45, 45).plot.hist(bins=90, figsize=(9, 4))
ax.axvline(0, color="black", lw=1, ls="--")
ax.set_xlabel("arr_delay − dep_delay (min); negative = time made up in the air")
ax.set_ylabel("number of flights")
ax.set_title("Time recovered (or added) between departure and arrival")
plt.tight_layout()
plt.show()
```

**Findings — departure delay all but determines arrival delay.**

- **Correlation is 0.971** → `dep_delay` explains **~94%** of the variance in `arr_delay` (r² ≈ 0.94). By the time a plane is airborne, how late it left already tells us almost exactly how late it will land — in the hexbin the mass hugs the `arr = dep` diagonal, sitting just below it.
- **`arr_delay − dep_delay` (time made up in the air):** mean **−5.5 min**, median **−7 min** — the typical flight *recovers* about 5–7 minutes aloft (schedule padding again). The spread is modest (std 13.9; 10th/90th pct = −20 / +10 min), with only rare extremes (min −502, max +654).

**Why this matters for our scope:** `dep_delay` is overwhelmingly the dominant signal — but it is **only known after the flight departs**. Because we've chosen a *before-departure* model (next section), `dep_delay` is off-limits, along with the other departure and arrival actuals. So this 0.94 is not a feature we get to use — it's the accuracy **ceiling we're deliberately giving up**, and it sets expectations: predicting arrival delay from the schedule alone is a genuinely harder problem, and the carrier / airport / time patterns that follow are the signal we actually have to work with.

### Modeling scope: predict *before departure*

**Decision:** the model forecasts at **scheduling / booking time**, before the aircraft pushes back. This fixes the legal feature set — anything known only once the flight is underway is off-limits.

**Available (known before departure):**
- Schedule: `month`, `day_of_month`, `day_of_week`, `fl_date`, `crs_dep_time`, `crs_arr_time`, `crs_elapsed_time`, `distance`
- Carrier & route: `op_unique_carrier`, `op_carrier_fl_num`, `origin`, `dest` (+ city / state names)

**Excluded (known only at/after departure — cannot be inputs):**
- `dep_time`, `dep_delay`, `taxi_out`, `wheels_off` — departure actuals
- `wheels_on`, `taxi_in`, `arr_time`, `actual_elapsed_time`, `air_time` — arrival / landing
- `carrier_delay`, `weather_delay`, `nas_delay`, `security_delay`, `late_aircraft_delay` — post-hoc cause attributions
- `cancelled`, `diverted`, `cancellation_code` — outcome flags

This is a **stricter cutoff than the inflight framing**: `dep_delay` — which alone explained ~94% of arrival delay — is now excluded. That's the point: the model must anticipate delay from the schedule, without seeing how late the flight actually left. So we predict `arr_delay` **directly** (the residual trick needs `dep_delay`), expect materially larger errors than the 0.97 baseline, and treat the carrier / airport / time patterns below as the **primary** signal. `year` (constant 2024) is dropped as dead.

### Who and when: delay by carrier, airport, and time

With `dep_delay` excluded (pre-departure model), these schedule-time attributes are the **primary** signal — *who* operates the flight and *when / where* it flies, all known at booking time. We break `arr_delay` down by:

- **carrier** — some airlines run tighter operations than others;
- **origin airport** — hub congestion and weather exposure vary widely;
- **time** — month (seasonal weather), day-of-week, and scheduled departure hour (delays cascade and build through the day).

These are the main leads for feature engineering. We report **median alongside mean**, since the heavy right tail pulls means upward.

```python

by_carrier = (flights.groupby("op_unique_carrier", observed=True)["arr_delay"]
              .agg(mean="mean", median="median", flights="count")
              .sort_values("mean", ascending=False)
              .round(1))
by_carrier
```

```python

by_origin = (flights.groupby("origin", observed=True)["arr_delay"]
             .agg(mean="mean", median="median", flights="count"))
busy = by_origin[by_origin["flights"] >= 10_000]
busy.sort_values("mean", ascending=False).head(15).round(1)
```

**Findings — carrier and origin airport both shift the delay, mostly in the tail.**

*Carrier* (mean, worst → best): **AA 15.3** and **F9 15.2** lead, then B6 / OH ~10; the most punctual are **YX −1.8** and **9E 1.7** (regionals), with **DL 3.7** notably strong for a major. But every carrier's **median is negative** (−1 to −11) — even the "worst" airline's *typical* flight lands early. So carriers differ far more in their **right tail** than in their typical flight: AA's mean (15.3) sits ~19 min above its median (−4), while YX / 9E run medians of −11. `HA` is the opposite shape — worst median (−1) but a low mean (4.2): Hawaiian's short inter-island hops cluster near zero, rarely very early or very late.

*Origin airport* (busy airports, worst mean): **MIA 14.3, DFW 13.9, CLT 13.9, SJU 13.8, IAH 12.4, FLL 12.3** — a mix of congested legacy hubs (DFW, CLT, IAH, ORD) and Florida / Caribbean airports (MIA, FLL, PBI, SJU) exposed to convective summer weather. Medians are again all negative (−2 to −7), so the airport effect, like the carrier effect, lives in the tail.

**For feature engineering:** carrier and origin (and likely `dest`) are genuinely informative and belong in the model — but because they move the *mean / tail* more than the median, they matter most for predicting the magnitude on delay-prone flights. Encode them as categoricals (or target-encode with a tail-aware statistic), and remember these raw means are inflated by the same heavy tail we saw in the target.

```python

overall = flights["arr_delay"].mean()
dep_hour = ((flights["crs_dep_time"] // 100) % 24).rename("dep_hour")

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

flights.groupby("month", observed=True)["arr_delay"].mean().plot(marker="o", ax=axes[0])
axes[0].set(title="by month", xlabel="month", ylabel="mean arr_delay (min)")

flights.groupby("day_of_week", observed=True)["arr_delay"].mean().plot(marker="o", ax=axes[1])
axes[1].set(title="by day of week (1=Mon)", xlabel="day_of_week", ylabel="")

flights.groupby(dep_hour, observed=True)["arr_delay"].mean().plot(marker="o", ax=axes[2])
axes[2].set(title="by scheduled departure hour", xlabel="hour", ylabel="")

for ax in axes:
    ax.axhline(overall, color="red", lw=1, ls="--")
    ax.grid(alpha=0.3)

fig.suptitle("Mean arrival delay across time")
plt.tight_layout()
plt.show()
```

**Findings — strong, usable time patterns (all known before departure).**

**Scheduled departure hour — the cleanest signal.** Mean delay is *negative* for early-morning departures (5–8 am arrive slightly early), then climbs almost monotonically through the day to a peak around **19–20:00 (~15–16 min)**, before dropping back late at night. This is the classic **delay cascade**: the first banks of flights leave on time, and delay accumulates through the day as aircraft and crews fall behind on their rotations. (The noisy blips at hours 1–4 are tiny-volume overnight slots — ignore them.) Departure hour is likely to be one of the most valuable engineered features.

**Month — two delay seasons.** A large **summer peak** (May–Aug, **July worst at ~18 min**) driven by convective thunderstorms and peak travel, plus a smaller **winter / holiday bump** (Dec–Jan). **Autumn (Sep–Nov) is the calmest** — Oct/Nov mean delay actually goes *negative*. February is surprisingly low.

**Day of week — the back half is worse.** **Friday is worst (~9.6 min)**, with Sunday and Thursday also above average; **Tuesday / Wednesday are smoothest (~4–5)**, and Saturday is notably light. This mirrors business + weekend travel demand.

**For feature engineering:** all three are known at scheduling time and each shifts the mean by ~10–18 min across its range — genuine signal. Engineer **`dep_hour`** (from `crs_dep_time`), **`month` / season**, and **`day_of_week`**. Because the hour and month effects are cyclical and non-monotonic, consider cyclical (sin/cos) encodings for linear models — tree models can use the raw integers directly.

### How far: delay vs flight distance

`distance` (and its near-twin `crs_elapsed_time`) is the last schedule feature to check. We bin flights into distance bands and look at mean / median `arr_delay` per band, then confirm how tightly `distance` and `crs_elapsed_time` move together — they encode almost the same thing, so we'd keep only one for a linear model.

```python
bins = [0, 250, 500, 750, 1000, 1500, 2000, 3000, 6000]
dist_band = pd.cut(flights["distance"], bins)
by_dist = (flights.groupby(dist_band, observed=True)["arr_delay"]
           .agg(mean="mean", median="median", flights="count")
           .round(1))
by_dist
```

```python
plot_df = by_dist.copy()
plot_df.index = plot_df.index.astype(str)

ax = plot_df[["mean", "median"]].plot(marker="o", figsize=(9, 4))
ax.axhline(flights["arr_delay"].mean(), color="red", lw=1, ls="--", label="overall mean")
ax.set_xlabel("distance band (miles)")
ax.set_ylabel("arr_delay (min)")
ax.set_title("Arrival delay by flight distance")
ax.tick_params(axis="x", rotation=30)
ax.grid(alpha=0.3)
ax.legend()
plt.tight_layout()
plt.show()
```

```python

flights[["distance", "crs_elapsed_time", "arr_delay"]].corr().round(3)
```

**Findings — distance barely matters (and duplicates `crs_elapsed_time`).**

- **`distance` ↔ `crs_elapsed_time` = 0.983** — essentially the same feature (longer flights take longer). Keep **`distance`** and drop `crs_elapsed_time` to avoid redundancy.
- **`distance` vs `arr_delay` = −0.006** (and `crs_elapsed_time` −0.009): **no linear relationship** — flight length on its own tells us almost nothing about arrival delay.
- The band table shows why that correlation is 0: the effect is **small and non-monotonic**. Mean delay is roughly flat (~6.5–8.2 min) from 0 to 1,500 miles, peaks around **1,000–1,500 mi (8.2)**, then *falls* for long-haul — **2,000–3,000 mi → 4.1**, **3,000+ → 3.1**. Long flights are a bit *less* delayed on average (more schedule padding and air time to absorb slack, fewer of them, less cascade exposure). Medians sit at ~−6 across every band, so distance barely moves the typical flight.

**For feature engineering:** drop `crs_elapsed_time` (collinear); keep `distance` as a **minor** feature — a tree can pick up the long-haul dip, but linearly it's near-useless. This confirms the EDA's ranking: **time-of-day ≫ carrier ≈ origin airport ≫ distance.**

### The tail: how often flights land 60+ minutes late

Means hide the tail — and the carrier and airport effects live there. So here we look directly at the **% of flights that arrive ≥ 60 minutes late** (the share of really bad delays), broken down by **carrier** and by **scheduled departure hour**.

```python

big = flights["arr_delay"] >= 60
pct = (big.groupby(flights["op_unique_carrier"], observed=True).mean()
         .mul(100).sort_values())

ax = pct.plot.barh(color="#2b6cb0", zorder=3)
for i, v in enumerate(pct):
    ax.text(v + 0.1, i, f"{v:.1f}%", va="center", fontsize=9)
ax.set(xlabel="% of flights 60+ min late", ylabel="carrier",
       title="% of flights arriving 60+ min late, by carrier")
ax.margins(x=0.12)
plt.tight_layout()
plt.show()
```

```python

dep_hour = ((flights["crs_dep_time"] // 100) % 24).rename("dep_hour")
p_by_hour = big.groupby(dep_hour, observed=True).mean().mul(100)

ax = p_by_hour.plot(marker="o", lw=2.2, color="#2b6cb0", zorder=3)
ax.fill_between(p_by_hour.index, p_by_hour.values, color="#2b6cb0", alpha=0.12)
ax.axhline(big.mean() * 100, color="#dd6b20", ls="--", lw=1.5, label="overall")
ax.set(xlabel="scheduled departure hour",
       ylabel="% of flights 60+ min late",
       title="% of flights arriving 60+ min late, by departure hour",
       xticks=range(0, 24, 2))
ax.legend()
plt.tight_layout()
plt.show()
```

### Congestion: does a busier airport-hour mean more delay?

Delays cascade partly because of **congestion** — many flights competing for the same runways and gates. As a proxy, for each flight we count how many departures share its **origin, date, and scheduled hour**, then track mean delay as that count grows. If delay rises with congestion, *"scheduled departures in the same airport-hour"* becomes a promising engineered feature — and it's fully known before departure.

```python
# How many flights share each flight's origin + date + scheduled hour (a congestion proxy).
# NB: the 3-key groupby over ~7M rows takes a few seconds.
dep_hour = (flights["crs_dep_time"] // 100) % 24
sched_in_hour = (flights.groupby(["origin", "fl_date", dep_hour], observed=True)["arr_delay"]
                 .transform("size"))

band = pd.cut(sched_in_hour, [0, 5, 10, 20, 30, 50, 10_000],
              labels=["1-5", "6-10", "11-20", "21-30", "31-50", "50+"])
by_cong = (flights.groupby(band, observed=True)["arr_delay"]
           .agg(mean="mean", median="median", flights="count").round(1))
by_cong
```

```python
ax = by_cong["mean"].plot.bar(color="#2b6cb0", width=0.72, zorder=3)
ax.axhline(flights["arr_delay"].mean(), color="#dd6b20", ls="--", lw=1.5, label="overall mean")
for i, v in enumerate(by_cong["mean"]):
    ax.text(i, v + 0.25, f"{v:.1f}", ha="center", va="bottom", fontsize=10)
ax.set(xlabel="scheduled departures in same origin-hour",
       ylabel="mean arr_delay (min)",
       title="Arrival delay rises with airport-hour congestion")
ax.tick_params(axis="x", rotation=0)
ax.margins(y=0.15)
ax.legend()
plt.tight_layout()
plt.show()
```

**Findings — congestion has only a modest effect.**

Grouping flights by how many departures share their **origin, date, and scheduled hour**, mean arrival delay drifts up with congestion but weakly and not cleanly: **6.8 → 5.7 → 7.2 → 7.1 → 7.4 → 8.5 min** from the quietest (1–5 departures) to the busiest (50+) airport-hours. The busiest slots (**50+ → 8.5 min**) are clearly the worst, but the 6–10 band dips, so it's noisy. The **median** trend is cleaner — a steady climb from **−8 min** (quietest) to **−5 min** (busiest), a ~3-minute shift.

**For feature engineering:** a *"scheduled departures in the same origin-hour"* count is a **mild** positive signal — cheap, known before departure, and worth including — but it's a supporting feature, not a driver. Much of its effect overlaps with departure-hour and airport, which already capture peak-period congestion.
